```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
testScoresDf = pd.read_csv("/content/test_scores.csv")
```

```python
testScoresDf
```

x = Math score and y = CS score for prediction purposes

```python
x = np.array(testScoresDf['math'])
y = np.array(testScoresDf['cs'])
```

```python
def gradient_descent(x,y):
  mCurrent = bCurrent = 0
  previousCost = 0
  currentCost = math.inf
  iterations = 1000000
  learning_Rate = 0.0002


  for i in range(iterations):
    previousCost = currentCost
    y_Predicted = mCurrent * x + bCurrent
    slopeM = -(2/len(x)) * sum(x * (y - y_Predicted))
    slopeB = -(2/len(x)) * sum(y - y_Predicted)
    mCurrent = mCurrent - learning_Rate * slopeM
    bCurrent = bCurrent - learning_Rate * slopeB
    yP = mCurrent * x + bCurrent
    currentCost = (1/len(x)) * sum((value**2) for value in (y - yP))
    if (math.isclose(previousCost,currentCost,rel_tol = 1e-20)):
      break
    print(f"mCurrent: {mCurrent}, bCurrent: {bCurrent}, currentCost: {currentCost}, iteration: {i}")
  
  return mCurrent, bCurrent
```

```python
gDResult = gradient_descent(x,y)
```

```python
print(f"Slope equals: {gDResult[0]}, intercept equals: {gDResult[1]}")
```
